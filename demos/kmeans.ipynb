```python
from sklearn import datasets
```

```python
# Fix import path
import sys
from os import path
sys.path.insert(0, path.abspath(path.join("..")))

from models.traditional.kmeans import KMeans
```

```python
N = 2000
K = 4
SEED = 122521
```

```python
X, Y, centers = datasets.make_blobs(n_samples=N,
                                    n_features=2,
                                    centers=K,
                                    shuffle=True,
                                    random_state=SEED,
                                    return_centers=True)
```

```python
kmeans = KMeans(X, K)
kmeans.fit()
```

```python
predictions = kmeans.predict(X)
centroids = kmeans._centroids
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 4))

plt.subplot(1, 2, 1)
plt.title('KMeans Clusters')
sns.scatterplot(x=X[:,0], y=X[:,1], hue=predictions, palette='Dark2')
sns.scatterplot(x=centroids[:,0], y=centroids[:,1], s=200)

plt.subplot(1, 2, 2)
plt.title('Actual Clusters')
sns.scatterplot(x=X[:,0], y=X[:,1], hue=Y, palette='Dark2')

plt.show()
```
